```python
import pyeeg
import pandas as pd
import numpy as np
from scipy.signal import welch
import matplotlib.pyplot as plt
import scipy.stats as stats

import os
import time
```

```python
index_channel = ["FP1", "AF3", "F3", "F7", "FC5", "FC1", "C3", "T7", "CP5", "CP1", "P3", "P7",
                 "PO3", "O1", "Oz", "Pz", "FP2", "AF4", "Fz", "F4", "F8", "FC6", "FC2", "Cz",
                 "C4", "T8", "CP6", "CP2", "P4", "P8", "PO4", "O2"]

#basic_channel = ["F3", "F4", "F7", "F8", "FC1", "FC2", "FC5", "FC6", "FP1", "FP2", "AF3", "AF4"]
c_index = [1,2,3,4,5,6,17,18,20,21,22,23] #DataFrame 1,,,~~

#0,1,2,3,4,5,16,17,19,20,21,22 <<real
```

```python
#csv column : calulated feature+label, row : video. /// csv file num = channel num


sampling_period = 1/128
sampling_frequency = 128

labels = ["HH", "LL"]
column_table = ["label", "mean", "std", "skewness", "kurotosis", "hjorth_mob", "hjorth_com",
                "hurst","dfa", "hfd","pfd", "MAX_PSD","MAX_freq","RMS"]

for c in c_index:
        
    rows = list(range(14)) #feature + Label
    listrow = list()
    
    for label in labels:
        direc = "/home/admin3/Downloads/DEAP_Arousal_Valence_group"
        dir = "/home/admin3/Downloads/DEAP_Arousal_Valence_group"+f"/{label}"

        for filename in os.listdir(dir):
            
            filepath = dir + f"/{filename}"
            start = time.time()
            fid = pd.read_csv(filepath, sep = ",", header=None)
            
            #fid.index = index_channel
            fid = fid.drop(0, axis=0)
            fid = fid.drop(0, axis=1)
            
            fid = fid.apply(pd.to_numeric)
            
            row = fid.iloc[c, 128*3:128*60]
            
            #Time-domain feature
            hjor_mob = pyeeg.hjorth(list(row))[0] #a one dimensional floating-point array representing a time series.
            hjor_com = pyeeg.hjorth(list(row))[1]
            DFA = pyeeg.dfa(row) 
            PFD = pyeeg.pfd(row) 
            HFD = pyeeg.hfd(list(row),8)
            hurst = pyeeg.hurst(row)

            ##statistic
            mean = np.mean(row)
            std = np.std(row)
            skewness = stats.skew(row)
            kurotosis = stats.kurtosis(row)
            
            #frequency-domain feature
            f, Pxx_den = welch(row, fs=128, nperseg=128)
            f_med, Pxx_den_med = welch(row, fs=128, nperseg=128, average = 'median')
        
            f, Pxx_spec = welch(row, fs=128, nperseg=128,scaling = 'spectrum')
    
            MAX_PSD = np.sqrt(Pxx_den.max())
            MAX_freq = np.argmax(Pxx_den)
            RMS = np.sqrt(Pxx_spec.max()) 

            print("OK")
            rows[0] = labels.index(label) #0=HH, 1=LL
            rows[1] = mean
            rows[2] = std
            rows[3] = skewness
            rows[4] = kurotosis
            rows[5] = hjor_mob
            rows[6] = hjor_com
            rows[7] = hurst
            rows[8] = DFA
            rows[9] = HFD
            rows[10] = PFD
            rows[11] = MAX_PSD
            rows[12] = MAX_freq
            rows[13] = RMS
            
            listrow.append(list(rows))
            print("channel:",index_channel[c-1],"Label:",label,"video:",filename,"is complete")
    
    ALLtable = pd.DataFrame(data = listrow, columns = column_table)

    ALLtable.to_csv(direc + f"/forML/channel_{index_channel[c-1]}.csv")
    print(f"channel_{index_channel[c-1]} calculation is complete: {time.time() - start}")

        




    
```
